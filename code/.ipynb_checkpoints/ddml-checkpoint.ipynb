```python
import numpy as np
import pandas as pd

from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor

import statsmodels.api as sm
```

```python
df = pd.read_excel("/Users/lmm/Documents/论文/ddml数据.xlsx")
df
```

```python
# =============== 1. 两维固定效应去均值，类似 reghdfe absorb(id YEAR) ===============

def twoway_demean(df, cols, id_col, time_col):
    """
    对 cols 这些变量做 双向固定效应 去均值：
    x_it - x_i. - x_.t + x_..
    """
    df = df.copy()
    for c in cols:
        grand_mean = df[c].mean()
        id_mean = df.groupby(id_col)[c].transform('mean')
        t_mean = df.groupby(time_col)[c].transform('mean')
        df[c + '_dm'] = df[c] - id_mean - t_mean + grand_mean
    return df
```

```python
# =============== 2. 双重机器学习（partialling-out）主函数 ===============

def dml_rf_partialling_out(df,
                           y_col,
                           d_cols,
                           z_cols,
                           id_col=None,
                           n_splits=5,
                           random_state=0):
    """
    双重机器学习估计  Y = theta * D + g(Z) + u
    第一阶段：随机森林估计 E[Y|Z], E[D|Z]
    第二阶段：对残差回归，得到 theta 及显著性（可选聚类稳健）
    """
    # 丢掉缺失
    used_cols = [y_col] + d_cols + z_cols
    data = df.dropna(subset=used_cols).copy()

    Y = data[y_col].values
    D = data[d_cols].values   # 可以是一列或多列
    Z = data[z_cols].values

    n, p_d = D.shape
    Y_tilde = np.zeros(n)
    D_tilde = np.zeros((n, p_d))

    kf = KFold(n_splits=n_splits, shuffle=True, random_state=random_state)

    for train_idx, test_idx in kf.split(Z):
        Z_train, Z_test = Z[train_idx], Z[test_idx]
        Y_train, Y_test = Y[train_idx], Y[test_idx]
        D_train, D_test = D[train_idx], D[test_idx]

        # ---------- 第一阶段：E[Y | Z] ----------
        rf_y = RandomForestRegressor(
            n_estimators=500,
            max_depth=None,
            min_samples_leaf=5,
            random_state=random_state,
            n_jobs=-1
        )
        rf_y.fit(Z_train, Y_train)
        m_hat = rf_y.predict(Z_test)

        # ---------- 第一阶段：E[D | Z] ----------
        if p_d == 1:
            rf_d = RandomForestRegressor(
                n_estimators=500,
                max_depth=None,
                min_samples_leaf=5,
                random_state=random_state,
                n_jobs=-1
            )
            rf_d.fit(Z_train, D_train.ravel())
            g_hat = rf_d.predict(Z_test).reshape(-1, 1)
        else:
            rf_d = MultiOutputRegressor(
                RandomForestRegressor(
                    n_estimators=500,
                    max_depth=None,
                    min_samples_leaf=5,
                    random_state=random_state,
                    n_jobs=-1
                )
            )
            rf_d.fit(Z_train, D_train)
            g_hat = rf_d.predict(Z_test)

        # ---------- 残差 ----------
        Y_tilde[test_idx] = Y_test - m_hat
        D_tilde[test_idx, :] = D_test - g_hat

    # ---------- 第二阶段：OLS 回归 Y残差 ~ D残差 ----------
    X_second = sm.add_constant(D_tilde)
    model = sm.OLS(Y_tilde, X_second)

    if id_col is not None:
        groups = data[id_col].values
        res = model.fit(cov_type='cluster', cov_kwds={'groups': groups})
    else:
        res = model.fit()

    return res, data, Y_tilde, D_tilde
```

```python
# =============== 3. 按你的设定调用 ===============

# 假设 df 是你的原始 DataFrame
# 1）先筛选 sample == 1
df_use = df[df["sample"] == 1].copy()
```

```python
# 2）做双向固定效应去均值（尽量贴近 reghdfe absorb(id YEAR)）
y_col_raw = "TFP_LP"
treat_col_raw = "investdid"   # 这里填你 Stata 里 `X` 的那一列名字

ctrls = [
    "lev", "roa", "size", "cash", "growth",
    "两职合一", "Big4", "独立董事占比", "股权集中度",
    "age2", "soe",
]

cols_to_demean = [y_col_raw, treat_col_raw] + ctrls

df_dm = twoway_demean(df_use, cols_to_demean, id_col="id", time_col="year")
```

```python
# 去均值后的列名：原名 + "_dm"
y_col = y_col_raw + "_dm"
d_cols = [treat_col_raw + "_dm"]
z_cols = [c + "_dm" for c in ctrls]
```

```python
# 3）跑 DML
res, data_used, Y_tilde, D_tilde = dml_rf_partialling_out(
    df_dm,
    y_col=y_col,
    d_cols=d_cols,
    z_cols=z_cols,
    id_col="id",          # 做按公司聚类稳健
    n_splits=5,
    random_state=2025
)

print(res.summary())
```
